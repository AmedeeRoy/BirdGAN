## split data

```python
trajs = data.trip.unique()
np.random.shuffle(trajs)
```

```python
nb_train = round(0.8 * len(data.trip.unique()))
nb_test = round(0.2 * len(data.trip.unique()))
```

```python
train_trajs = trajs[0:nb_train]
test_trajs = trajs[nb_train:(nb_train+nb_test)]
```

```python
data_train = data[data.trip.isin(train_trajs)]
data_test = data[data.trip.isin(test_trajs)]
```

```python
(nb_train, nb_test)
```

## Data loader

```python
import torch
from torch import nn
from torch import optim
from torch.utils.data import DataLoader, Dataset
```

```python
class TrajDataSet(Dataset):
    def __init__(self,  df, window, transform=None):
        self.df = df.set_index(np.arange(len(df))) #reorder idx
        self.window = window
        self.start_idx = np.where([self.df.trip[i]==self.df.trip[i+self.window] for i in range(len(self.df)-self.window)])[0]
        self.transform = transform

    def __len__(self):
        return len(self.start_idx)

    def __getitem__(self, idx):

        i = self.start_idx[idx]

        # select coordinates
        coord = self.df.loc[i:i+self.window, ('lon_std', 'lat_std', 'prediction')]
        x = np.array(coord)[:-1,:]
        y = np.array(coord)[-1,:]

        
        sample = (x, y)
        if self.transform:
            sample = self.transform(sample)

        return sample

class ToTensor(object):
    """Convert ndarrays in sample to Tensors."""

    def __call__(self, sample):
        (x,y) = sample
        return (torch.FloatTensor(x), torch.FloatTensor(y))
```

```python
# hyperparameters
batch_size = 64
window = 20

## reduce size dataset
train_set = TrajDataSet(data_train, window = window, transform=ToTensor())
train_loader = DataLoader(train_set, batch_size=batch_size, num_workers = 0, shuffle = True, drop_last=True)

test_set = TrajDataSet(data_test, window = window, transform=ToTensor())
test_loader = DataLoader(test_set, batch_size=batch_size, num_workers = 0, shuffle = True, drop_last=True)
```

```python
train_set[0]
```

## model LSTM

```python
class LSTM(torch.nn.Module):
    def __init__(self,n_features,seq_length):
        super(LSTM, self).__init__()
        self.n_features = n_features
        self.seq_len = seq_length
        self.n_hidden = 200 # number of hidden states
        self.n_layers = 10 # number of LSTM layers (stacked)
    
        self.l_lstm = torch.nn.LSTM(input_size = n_features, 
                                 hidden_size = self.n_hidden,
                                 num_layers = self.n_layers, 
                                 batch_first = True)
        # according to pytorch docs LSTM output is 
        # (batch_size,seq_len, num_directions * hidden_size)
        # when considering batch_first = True
        self.l_linear = torch.nn.Linear(self.n_hidden*self.seq_len, 3)
        
    
    def init_hidden(self, batch_size):
        # even with batch_first = True this remains same as docs
        hidden_state = torch.zeros(self.n_layers,batch_size,self.n_hidden)
        cell_state = torch.zeros(self.n_layers,batch_size,self.n_hidden)
        self.hidden = (hidden_state, cell_state)
    
    
    def forward(self, x):        
        batch_size, seq_len, _ = x.size()
        
        lstm_out, self.hidden = self.l_lstm(x,self.hidden)
        # lstm_out(with batch_first = True) is 
        # (batch_size,seq_len,num_directions * hidden_size)
        # for following linear layer we want to keep batch_size dimension and merge rest       
        # .contiguous() -> solves tensor compatibility error
        x = lstm_out.contiguous().view(batch_size,-1)
        return self.l_linear(x)
```

```python
model = LSTM(3,window)
model.init_hidden(batch_size)
```

## training

```python
# get sample
x, y = next(iter(train_loader)) 

out = model(x)

# Loss and score
learning_rate = 0.001
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
criterion = nn.MSELoss()
criterion(out, y)
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# Train the model
nb_epoch = 5

global_loss_train = []
global_loss_test = []
    
for epoch in range(nb_epoch):
    optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
    i = 0
    for batch, (x, y) in enumerate(train_loader):
        i+=1
        # send to GPU
        x, y = x.to(device), y.to(device)

        # Run the forward pass
        model.init_hidden(batch_size)
        out = model(x)
        loss = criterion(out, y)
        
        # Backprop and perform optimisation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

    ### Evaluation + test every epoch
    model.eval()
    with torch.no_grad():      
        j = 0
        # evaluation
        list_loss_train = []
        for batch, (x, y) in enumerate(train_loader):
            j+= 1
            # send to GPU
            x, y = x.to(device), y.to(device)

            # Run the forward pass
            out =  model(x)
            loss = criterion(out, y)
            list_loss_train.append(loss.item())
        
        train_loss = np.mean(list_loss_train)
            
        k = 0
        # test
        list_loss_test = []
        for batch, (x, y) in enumerate(test_loader):
            k+= 1
            # send to GPU
            x, y = x.to(device), y.to(device)

            # Run the forward pass
            out =  model(x)
            loss = criterion(out, y)
            list_loss_test.append(loss.item())

        test_loss = np.mean(list_loss_test)

    print('Epoch [{}/{}] -------------------------------------------------------------------------------------'
          .format(epoch+1, nb_epoch))
    print('Train Loss: {} Test Loss: {}'
            .format(round(train_loss, 2), round(test_loss, 2)))
    model.train()
    
    global_loss_train.append(train_loss)
    global_loss_test.append(test_loss)

    #### STOP WHEN test DECREASE
    if epoch>1:
        if global_loss_test[-2] < global_loss_test[-1]:
            break
```
